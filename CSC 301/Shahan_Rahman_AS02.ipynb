```python
#Shahan Rahman Assignment two 
import math
import time

def naiveSum(num):
    sn = 0
    temp = 0
    for i in range(num):
        temp += math.pi
        sn = temp
    return sn

def comSum(num):
    sn = 0
    com = 0
    for i in range(num):
        y = math.pi - com
        t = sn + y
        com = (t-sn) - y
        sn = t
    return sn

def exact(num):
    return (num * math.pi)

for x in range(6, 10):
    print("For 10^" + str(x))
    startTime = time.time()
    num = int(math.pow(10,x))
    naive = naiveSum(num)
    print("Sum by naive summation is : ")
    print(naive)
    print("Sum by compensated summation is : ")
    compensated = comSum(num)
    print(compensated)
    print("Exact value : ")
    exactVal = exact(num)
    print(exactVal)
    print("")
    print("Relative error of naive : ")
    print(naive - exactVal)
    print("Absolute error of naive : ")
    print(math.fabs(naive - exactVal))
    print("")
    print("Relative error of compensated : ")
    print(compensated - exactVal)
    print("Absolute error of compensated : ")
    print(math.fabs(compensated - exactVal))
    print("")
    endTime = time.time()
    print("Total time : ")
    print(endTime-startTime)
    print("")
    print("")
    print("")
    print("")
    print("")
print("In my program the compensated summation is the same value as the exact value therefore the \n error of the compensated should be zero.")
print("This does match what we see in lecture because compensated or kahan summation is more accuracy thank the naive method.")
```
