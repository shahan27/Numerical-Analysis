```python
import math
from math import *
from sympy.functions import sin, cos

'''
Shahan Rahman
Analysis:
I used the Taylor series, forward and backward Euler, and three point midpoint to calculate sin(x).
When the value of n increases the accuracy increases too. But for the foward and the backward method 
is not the best for the inverse sin functions since the absolute and relative error is high compared to the other 
two methods. Therefore we can conclude with more points the accuracy increases too. 
'''



def f(x):
    return cos(x)

def taylorFun(x,n):
    temp = x
    cos = temp
    sign = 1
    for i in range(1,n):
        sign = sign *(-1)
        mult = sign*(x**(2*i+1))/(math.factorial(2*i+1))
        cos += mult
    return cos

def forwardEuler(func, a):
    h = 0.00000001
    top = func(a + h) - func(a)
    bottom = h
    slope = top/bottom
    return float("%.16f"% slope)

def backwardEuler(func, a):
    h = 0.00000001
    top = func(a) - func(a - h)
    bottom = h
    slope = top/bottom
    return float("%.16f"% slope)


def three_point_midpoint(func, a):
    h = 0.00000001
    top = func(a+h) - func(a - h)
    bottom = 2*h
    slope = top/bottom
    return float("%.16f"% slope)

#First Method Taylor Series 
measureTaylorCos = taylorFun(1,500) #500 is for a more acc result
ExactCos = float(cos(1))
print("Taylor series for finding cos(1) = ", measureTaylorCos)    
print("Exact value of cos(1) =", ExactCos)
absTaylorCos = measureTaylorCos - ExactCos
relativeTaylorCos = (measureTaylorCos - ExactCos)/ExactCos
print("The absolute Error = ", absTaylorCos)
print("The relative Error = ", relativeTaylorCos)

#Second Method Forward Euler
measureValue = forwardEuler(f,1)
ExactValue = float(cos(1))
absError = measureValue - ExactValue
relativeError = (measureValue - ExactValue)/ExactValue
print("\n\n\nThe exact value of cos'(x) = ",ExactValue)
print("Forward Euler formula = ", measureValue)
print("The absolute Error  = ", absError)
print("The relative Error =", relativeError)

#Third Method Backward Euler
measureValue1 =backwardEuler(f,1)
absError1 = measureValue1 - ExactValue
relativeError1 = (measureValue1 - ExactValue)/ExactValue
print("\n\n\nThe exact value of cos'(x) = ",ExactValue)
print("Backward Euler formula = ", measureValue1)
print("The absolute Error  = ", absError1)
print("The relative Error =", relativeError1)

#Fourth Method Three point midpoint
measureValue2 = three_point_midpoint(f,1)
absError2 = measureValue2 - ExactValue
relativeError2 =  (measureValue2 - ExactValue)/ExactValue
print("\n\n\nThe exact value of cos'(x) = ",ExactValue)
print("Three-point-midpoint formula = ", measureValue2)
print("The absolute Error  = ", absError2)
print("The relative Error =", relativeError2)




```
