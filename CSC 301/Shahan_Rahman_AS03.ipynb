```python
import sympy as sy
import numpy as np
from sympy.functions import sin, cos, exp, log
import math
import matplotlib.pyplot as plt

x = sy.Symbol('x')
def taylor(fun, a, n):
    i = 0
    polynomial = 0
    while i <= n:
        polynomial = polynomial + (fun.diff(x,i).subs(x, a))/(math.factorial(i))*(x - a)**i
        i += 1
    return polynomial


f = sin(x)
n = [1, 2, 3, 4, 8, 16, 32, 64, 128, 256]
def getFunctions(nums,f):
    functionList = []
    for i in nums:
        func = taylor(f,0,i)
        functionList.append(func)
    return functionList

xlist = [1/32, 1/16, 1/8, 1/4, 1/2, 1, 2, 4]
FunctionList = getFunctions(n,f)
def getAnswer(FunctionList):
    print("\n SIN: \n")
    for temp in range(0, len(n)):
        print("===============================================================================")
        print('when n =' + str (n[temp]))
        print("===============================================================================")
        for i in xlist:
            Answer = FunctionList[temp].subs(x,i)
            measure = Answer
            actual = sin(i)
            absError = measure - actual
            absError = float('%.15f' % absError)
            relativeError = (measure - actual) / actual
            relativeError = float('%.15f' % relativeError)
            Answer = float('%.20f' % Answer)
            print('if x ='+ str(i), '---',Answer,"--Actual value is: ", actual, "--absErr is: ", absError,"--relativeErr is: ",relativeError)
    return 
answerList = getAnswer(FunctionList)
print("#####################################################################################################################")
print("#####################################################################################################################")


def graphOfSin(f, x_lims = [-15, 15], y_lims = [-15, 15], npoints = 700):
    x1 = np.linspace(x_lims[0], x_lims[1], npoints)
    y1 = []
    n = [1, 2, 3, 4, 8, 16, 32, 64,128,256]
    for j in n:
        func = taylor(f, 0, j)
        for k in x1:
            y1.append(func.subs(x,k))
        plt.plot(x1, y1, label = 'n = '+ str(j))
        y1 = []
    plt.plot(x1, np.sin(x1), label = 'f = sin(x)')  
    fig = plt.gcf()
    fig.set_size_inches(14.5, 10.5, forward=True)
    plt.xlim(x_lims)
    plt.ylim(y_lims)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.title('Taylor polynomial : f = sin(x)')
    plt.show()
    
graphOfSin(f)



print("#####################################################################################################################")
print("#####################################################################################################################")




g = cos(x)
FunctionList1 = getFunctions(n,g)
def getAnswer(FunctionList1):
    print("\n COS: \n")
    for temp in range(0, len(n)):
        print("===============================================================================")
        print('when n ='+str(n[temp]))
        print("===============================================================================")
        for i in xlist:
            Answer = FunctionList1[temp].subs(x,i)
            measure = Answer
            actual = cos(i)
            absError = measure - actual
            absError = float(absError)
            relativeError = (measure - actual) / actual
            relativeError = float(relativeError)
            Answer = float(Answer)
            print('if x ='+ str(i), '---',Answer,"--Actual value is: ", actual, "--absErr is: ", absError,"--relativeErr is: ",relativeError)
    return 
answerList = getAnswer(FunctionList1)

print("#####################################################################################################################")
print("#####################################################################################################################")


def graphOfCos(f, x_lims = [-15, 15], y_lims = [-15, 15], npoints = 700):
    x1 = np.linspace(x_lims[0], x_lims[1], npoints)
    y1 = []
    n = [1, 2, 3, 4, 8, 16, 32, 64,128,256]
    for j in n:
        func = taylor(f, 0, j)
        for k in x1:
            y1.append(func.subs(x,k))
        plt.plot(x1, y1, label = 'n = '+ str(j))
        y1 = []
    plt.plot(x1, np.cos(x1), label = 'f = cos(x)')     # give a cos(x) graph 
    fig = plt.gcf()
    fig.set_size_inches(14.5, 10.5, forward=True)
    plt.xlim(x_lims)
    plt.ylim(y_lims)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.title('Taylor polynomial : f = cos(x)')
    plt.show()

graphOfCos(g)

print("#####################################################################################################################")
print("#####################################################################################################################")



h = exp(x)
FunctionList2 = getFunctions(n,h)
def getAnswer(FunctionList2):
    print("\n Exp: \n")
    for temp in range(0, len(n)):
        print("===============================================================================")
        print('when n ='+str(n[temp]))
        print("===============================================================================")
        for i in xlist:
            Answer = FunctionList2[temp].subs(x,i)
            measure = Answer
            actual = exp(i)
            absError = measure - actual
            absError = float(absError)
            relativeError = (measure - actual) / actual
            relativeError = float(relativeError)
            Answer = float(Answer)
            print('if x ='+ str(i), '---',Answer,"--Actual value is: ", actual, "--absErr is: ", absError,"--relativeErr is: ",relativeError)
    return 
answerList = getAnswer(FunctionList2)

print("#####################################################################################################################")
print("#####################################################################################################################")


def graphOfexp(f, x_lims = [-15, 15], y_lims = [-15, 15], npoints = 700):
    x1 = np.linspace(x_lims[0], x_lims[1], npoints)
    y1 = []
    n = [1, 2, 3, 4, 8, 16, 32, 64,128,256]
    for j in n:
        func = taylor(h, 0, j)
        for k in x1:
            y1.append(func.subs(x,k))
        plt.plot(x1, y1, label = 'n = '+ str(j))
        y1 = []
    plt.plot(x1, np.exp(x1), label = 'f = exp(x)')     # give a exp(x) graph 
    fig = plt.gcf()
    fig.set_size_inches(14.5, 10.5, forward=True)
    plt.xlim(x_lims)
    plt.ylim(y_lims)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.title('Taylor polynomial : f = exp(x)')
    plt.show()
    
graphOfexp(h)








print("#####################################################################################################################")
print("#####################################################################################################################")


L = log(1-x)
xlistforLog = [1/32, 1/16, 1/8, 1/4, 1/2]
FunctionList3 = getFunctions(n,L)
def getAnswer(FunctionList3):
    print("\n Log: \n")
    for temp in range(0, len(n)):
        print("===============================================================================")
        print('when n ='+str(n[temp]))
        print("===============================================================================")
        for i in xlistforLog:
            Answer = FunctionList3[temp].subs(x,i)
            measure = Answer
            actual = log(1-i)
            absError = measure - actual
            absError = float(absError)
            relativeError = (measure - actual) / actual
            relativeError = float(relativeError)
            Answer = float(Answer)
            print('if x ='+ str(i), '---',Answer,"--Actual value is: ", actual, "--absErr is: ", absError,"--relativeErr is: ",relativeError)
    return 
answerList = getAnswer(FunctionList3)


print("#####################################################################################################################")
print("#####################################################################################################################")


def graphOfLog(f, x_lims = [-15, 15], y_lims = [-15, 15], npoints = 700):
    x1 = np.linspace(x_lims[0], x_lims[1], npoints)
    y1 = []
    n = [1, 2, 3, 4, 8, 16, 32, 64,128,256]
    for j in n:
        func = taylor(f, 0, j)
        for k in x1:
            y1.append(func.subs(x,k))
        plt.plot(x1, y1, label = 'n = '+ str(j))
        y1 = []
    plt.plot(x1, np.log(1-x1), label = 'f = log(1-x)')     # give a log(1-x) graph 
    fig = plt.gcf()
    fig.set_size_inches(14.5, 10.5, forward=True)
    plt.xlim(x_lims)
    plt.ylim(y_lims)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.title('Taylor polynomial : f = log(1-x)')
    plt.show()

    
graphOfLog(L)








print("#####################################################################################################################")
print("#####################################################################################################################")

```
